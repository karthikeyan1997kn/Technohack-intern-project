```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = r"C:\Users\KARTHIK\Downloads\Iris.csv"
data = pd.read_csv(file_path)

# Choose the column to visualize
column_to_visualize = 'Species'

# Count the occurrences of each category
category_counts = data[column_to_visualize].value_counts()

# Plotting the bar chart
category_counts.plot(kind='bar', color='purple', edgecolor='black')

# Adding labels and title
plt.xlabel(column_to_visualize)
plt.ylabel('Count')
plt.title(f'Bar Chart of {column_to_visualize}')

# Show the plot
plt.show()
```
